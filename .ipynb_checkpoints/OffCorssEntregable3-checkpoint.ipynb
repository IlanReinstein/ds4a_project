````
# Import the Offcorss data from all available files
def import_campaigns_offcorss(directory='/content/drive/My Drive/DS4A/PROJECT/DATA_OFFCORSS/'):
  '''This function retrieves all the information from campaigns, filters by moda and extracts 
  the important features to have a more consolidated dataframe'''
  columns_of_interest = ['AÑO', 'CAMPAÑA', 'REGION', 'DESCRIPICION_ZONA', 'NUMERO_ZONA', 'CLASIFICACION', 'ESTRATEGIA', 'GRUPO_ARTICULO', 'TIPO PRENDA', 'NUMERO_PEDIDO', 'REFERENCIA', 'COLOR_COMERCIAL',
                         'CODIGO_COLOR', 'TALLA', 'SKU_PLU', 'PRECIO_CATALOGO', 'COSTO', 'DDA UND', 'FAC UND', 'DDA $$', 'FAC $$', 'MARGEN', 'NIVEL_SERVICIO', 'NUMERO_DE_PEDIDOS', 'NPU_D', 'UPE_D', 
                         'ORDEN_PROMEDIO', 'DOCUMENTOCLIENTE'] # Many of these columns can be dropped
  list_campaigns = []
  for filename in os.listdir(directory):
      if 'CAMPA' in filename:
        dCampaign = pd.read_excel(os.path.join(directory, filename))
        list_campaigns.append(dCampaign)
  dfCampaigns = pd.concat(list_campaigns)[columns_of_interest]
  dfCampaigns.columns = dfCampaigns.columns.str.replace(' ', '_')
  # Filter the campaigns
  moda = dfCampaigns[dfCampaigns['CLASIFICACION']== 'Moda']
  zero_demand = (moda['DDA_UND'] == 0) & (moda['FAC_UND'] != 0)
  moda_campaign = moda[~zero_demand].groupby(['CAMPAÑA', 'SKU_PLU', 'TIPO_PRENDA',
                                        'GRUPO_ARTICULO']).agg({'DDA_UND':'sum',
                                          'FAC_UND':'sum',
                                          'PRECIO_CATALOGO':'mean',
                                          'UPE_D':'mean'}).reset_index()
  return moda_campaign
````

**Project Summary**: For this project, our goal is to generate demand predictions for a specific sales channel of the client. As we mentioned in the Project Scoping, there is a specific sales channel on which will focus our attention, that is direct sales, which are completed through sales-persons face-to-face with clients via a physical catalog. Such catalog is launched 12 times in a year and it contains a collection per catalog from which clients can choose from. We will be focusing on a single campaign: "Moda". This campaign changes for every catalog, hence the challenge of making accurate and asertive prediction on the demand.

Each sales-person places the order to Off-Corss without having true ground knowledge of the availability of inventory, therefore, the demand on these orders is not always met, and our client and the sales people incur in significant lossess. In order to minimze these losses our client's planning team uses an ad-hoc procedure based on prior catalog sales to estimate the number of units that should be put into production.

**Data Generating Process** The way Off-Corss makes predictions is the following: There are multiple stages at which they need to do a forecast in order to plan the production. 

 - Stage P1 (First forecast): The planning team receives a sketch of the references that will be displayed in a catalog at least six months in advanced (prior to sales). These are merely drawings on a spreadsheet which are assessed by the team. Taking specific features of the sketch, they manually look for a similar item/reference from the previous catalog/collection and based on the performance metrics like UPE they make an estimate of how many units should be produced of such new references. For example, a sketch for a red T-Shirt is received. The team looks for the most similar product on the sales history, uses the UPE to assign a weight, and makes an estimate for how many of these references will be needed in production. 

- Stage P2 (Second forecast): A second display of the item is sent to the plannign team to reassess or correct their initial guess. At this stage, they will already know how will the piece actually look in the catalog (graphical proposal) and they'll be able to extract the qualitative features such as page of the item, any stamp that the item may have, if it's modeled or not, the zoom of the image, and other information not previously available to them. At this point they redo the calculations and send a new estimate to the production team. This takes place approximately two months prior to sales).

- Stage P3 (Final forecast): At this point not many changes to their initial guess take place. Around 5% to 10% of corrections to their estimate are done.

**NOTE:** We are interested only on P1. If the client manages to imporove their assertivenes in this early stage, then number modifications to their guesses in subsequent stages will be reduced. 

**Data**: Out client will provide us two different data sources extracted at the end of a single catalog, i.e., one of the twelve that produces in the year. The central source is related to **Campaña** since it contains information of history of sales, such as the region of the order, date, specific reference features (color, size, price, collection), demand (how many units were ordered), among other. The second sources **Matriz** hold information on how the specific item was displayed in the catalog so we can find variables like page number, location on page, if the item has prints on it, the zoom on the item, and other information which is not entirely structure or correlated with our client's current forecasting approach. 

At this point we already have data for four (4) campaigns along with their respective qualitative data matrices. The goal of this second EDA is to have a closer look at the relevant variables that may be predictive of demand according to our client. The second is to test some of the common knowledge acquired empirically by the Planning team to verify that including qualitative information about the the display of references does actually improve their performance metrics.

**Metrics:** Some important metrics of interest to us and to our client are the UPE for each reference and the assertiveness. These metrics will prove to be tremendously useful when validating our model and testing it against the current workflow as we have already relevant test statistics to compare our predictions and know if we are adding value.

UPE is defined as the number of orders for product _p_ in time _t_ divided by the total number of orders at _t_.

Assertiveness is currently computed as the logarithm of the ratio between the true demand and the predicted demand. Currently the assertiveness for the 'Moda' channel is around 30%.

Before we look at any regression model, we have to look at our response variable and see if its somewhat normally distributed. Given that the demand is zero for many references, fitting a linear regression on this scale will not be very helpful, therefore we need to transform our data using the logarithm to insure all the assumptions are satisfied.

````
moda_agg['logDDA'] = np.log(moda_agg['DDA UND'] + 1)
moda_agg['logDDA'].hist()
````

As we see, there are many references for which there was zero demand (and zero invoices), therefore our data is zero inflated and will need proper care for our models. For now, we can get rid of this observations in order to make some tests.

````
zero_refs = moda_agg['logDDA'] == 0
moda_nonzero = moda_agg[~zero_refs]
````

The first thing we'll want to test is whether price is a predictive factor of demand.

````
import statsmodels.formula.api as smf
import statsmodels.api         as sm
model_price = smf.ols(formula = "logDDA ~ PRECIO_CATALOGO", 
                 data = moda_nonzero).fit()
print(model_price.summary())
````

As we can see here, the effect of price on demand is very small and significant. This does not mean that this variable is irrelevant but instead we are seeing that it is very weakly correlated with the intention of purchase. Specifically, an increase in price is associated with an almost negligible change in the number of requested units.

---------

So far, we have only looked at the relationship between price and demand, but we'd like to know if there are differences at a higher level of grouping of these references, specifically, ITEM GROUP.

````
 # Fit a regression between the log of demand and each item group.
 # Note that we have nthe intercept will be associated with the category "BODY"
model_group = smf.ols(formula = "logDDA ~ C(GRUPO_ARTICULO)", 
                 data = moda_nonzero).fit()
print(model_group.summary())
````

From the regression above, we can take the categordy associated with most sales and references "CAMISETA" and compare that to the baseline "BODY".

Notice that the Group associate with higer number of demanded units is the baseline BODY. This just means that these are the most requested items, and not specifically the one that sell the most. The rest of categories, are associated with smaller numbers on the demand.

One question we have, which is not of priority for us at this early stage is whether some references or types of items are sold differently by region. This is a problem for us, since getting the region information is complicated, however, we could look at historical data to test our hypothesis. Depending on the outcome of this next step, we'll be able to determine if we require regional information on salespersons for our model.

````
region_moda = moda_clean.groupby(['PLU', 'TIPO PRENDA', 
                               'GRUPO_ARTICULO', 'REGION']).agg({'DDA UND':'sum',
                                          'FAC UND':'sum',
                                          'PRECIO_CATALOGO':'mean', 
                                          'UPE_D':'mean'}).reset_index()
                              
region_moda['logDDA'] = np.log(region_moda['DDA UND'] + 1)
regions = region_moda['REGION'].unique()
````

````
model_region = smf.ols(formula = "logDDA ~ C(REGION)", 
                 data = region_moda).fit()
print(model_region.summary())
````

````
# Perform ANOVA test to validate if there is at least one of the averages that differs
aov_table = sm.stats.anova_lm(model_region, typ=2)
aov_table
````

````
# Finally, do independent t-test for each group to look for differences
from scipy import stats
gr = region_moda.groupby('REGION')
for i in range(len(regions)):
    for j in range(i+1, len(regions)):
        group1 = gr[["DDA UND"]].get_group(regions[i])
        group2 = gr[["DDA UND"]].get_group(regions[j])
        stat, pvalue = stats.ttest_ind(group1, group2, equal_var = False)
        print(regions[i] + " vs. " + regions[j] + "\t" + str(stat[0]) + "\t" + str(pvalue[0]))
````

From the above multiple t-test, ANOVA and regression we see that there are indeed differences in demand by region. We have not yet determined if there are differences by different type of articles. This is an interaction effect that will need further analysis.

The previous models are a simple version of an attempt to explain demand, limited to these unstructured variables. It can be seen that, even if both categories are taken into account, the model does not adequately fit the demand values. Practically, these categories (for the most part) do not explain the behavior of demand at all. Thus, below is a more complex model that can correct these imperfections.

````
temp = cat7[['DDA UND','PRECIO_CATALOGO', '# PÁGINA', 'NUM_ APARICIONES', 'PESO_ EXHIBICIÓN']]
temp['# PÁGINA'] = pd.to_numeric(temp['# PÁGINA'])
temp = pd.get_dummies(temp, columns=['NUM_ APARICIONES', 'PESO_ EXHIBICIÓN'])
f,ax = plt.subplots(figsize=(18, 18))
sns.heatmap(temp.corr(), annot=True, linewidths=.5, fmt= '.1f',ax=ax)
plt.show()
````

````
#Definición del modelo
temp = cat7.copy()
temp.rename(columns={'NUM_ APARICIONES':'NUM_APARICIONES', 'DDA UND': 'DDA_UND', 
                     'PESO_ EXHIBICIÓN': 'PESO_EXHIBICIÓN', 'TIPO PRENDA': 'TIPO_PRENDA'}, inplace = True)

modelo1 = 'np.log(DDA_UND+1)~NUM_APARICIONES + PRECIO_CATALOGO + PESO_EXHIBICIÓN + CODIGO_COLOR + TIPO_PRENDA +EXPOSICIÓN '

#Entrenamiento del modelo con la librería 'statsmodels'
lm1 = sm.ols(formula = modelo1, data = temp).fit()
print(lm1.summary())
````

With this more complex model, we have achieved a better fit for estimating the quantities demanded. Thus, we see that many of the existing categories in our database can be useful when making predictions.
